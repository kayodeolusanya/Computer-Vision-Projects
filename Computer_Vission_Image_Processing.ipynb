```python
import cv2
```

```python
#Loading the image using cv2
pics = cv2.imread('/content/RM.webp')
pics
```

```python
type(pics)
```

```python
pics.shape
```

```python
#visualize the data
#Opencv uses a different color channel called BGR(BLUE. GREEN AND BLUE)
plt.imshow(pics);
```

```python
#Converting pics from BGR to RGB
pics_c = cv2.cvtColor(pics, cv2.COLOR_BGR2RGB)
plt.imshow(pics_c);
```

```python
#Resizing the Pix
r_pics = cv2.resize(pics_c,(450,220))
plt.imshow(r_pics)
```

```python
#Flipping the image - 0, flips pic horizontally
f_pics = cv2.flip(pics_c, 0)
plt.imshow(f_pics)
```

```python
#Flipping the image - 0, flips pic horizontally
f_pics = cv2.flip(pics_c, -1)
plt.imshow(f_pics)
```

```python
#Create a blanck canvas
blank = np.zeros(shape=(1000,1000,3), dtype=np.int16)
blank
```

```python
plt.imshow(blank)
```

```python
#Drawing a rectangle
cblank = blank.copy()
```

```python
#PT1= TOP LEFT CORNER, PT2=BOTTOM RIGHT CORNER (Width, Height)
cv2.rectangle(cblank, pt1=(200,200),pt2=(400,400), color=(255,0,0),thickness=5)
plt.imshow(cblank)
```

```python
cv2.rectangle(cblank, pt1=(200,200),pt2=(400,400), color=(255,0,0),thickness=5)
cv2.rectangle(cblank, pt1=(620,610),pt2=(820,810), color=(0,255,0),thickness=5)
plt.imshow(cblank)
```

```python
#Drawing a Circle
cv2.circle(cblank, center=(300,300),radius=(90), color=(0,0,255),thickness=5)
plt.imshow(cblank)
```

```python
cv2.circle(cblank, center=(700,700),radius=(80), color=(100,255,100),thickness=5)
plt.imshow(cblank)
```

```python
#To have a solid color
cv2.circle(cblank, center=(700,700),radius=(80), color=(100,255,100),thickness=-1)
plt.imshow(cblank)
```
